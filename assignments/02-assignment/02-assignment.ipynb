1.  What is the fundamental problem of causal inference, and how does the potential outcomes framework help us understand it? 

#### The fundamental problem of causal inference is the inability to observe both theoretical outcomes of being treated or untreated. It is impossible to treat someone, and leave them untreated. Likewise, it is impossible to have someone who was untreated, be treated at the same time, and observe their outcomes. 

#### The potential outcomes framework addresses this by defining two possible outcomes: one if the individual receives treatment and one if they do not, but only one is observed. To estimate causal effects, methods like randomization in experiments or statistical techniques (e.g., matching, instrumental variables) in observational studies to approximate counterfactuals.

#### No, they are not necessarily the same. E[Yi(1)] represents the expected outcome if everyone in the population were treated, while E[Yi(1)∣T=1] is the expected outcome for only those who actually received the treatment. Therefore, E[Yi(1)∣T=1] does not necessarily generalize to the entire population.


3.  Explain the concept of selection bias. How does it affect our ability to estimate causal effects? 

#### Selection bias occurs when the individuals who receive treatment differ systematically from those who do not, leading to non-random assignment. This bias distorts causal estimates because the observed differences in outcomes may be due to preexisting characteristics rather than the treatment itself. 

#### As a result, comparing treated and untreated groups directly may not reveal the true causal effect but instead reflect underlying differences between them. To mitigate selection bias, researchers use methods like randomization, matching, or instrumental variables to approximate a valid counterfactual.

4.  What is the Stable Unit Treatment Value Assumption (SUTVA), and why is it important for causal inference? Give an example of a situation where SUTVA might be violated (not included in the slides).

#### The Stable Unit Treatment Value Assumption (SUTVA) states that (1) an individual’s potential outcomes are unaffected by the treatment status of others (no interference) and (2) there is only one version of the treatment (no hidden variations). This assumption is crucial for causal inference because it ensures that each unit's outcome depends only on its own treatment status, allowing for a clear comparison between treated and untreated groups. If SUTVA is violated, estimated causal effects may be misleading since external influences or treatment variations can distort the observed relationships. 

#### An example of a violation is peer effects in education, where a student’s academic performance might be influenced not only by their own access to tutoring (treatment) but also by whether their classmates receive tutoring, creating interference.

5.  What is a "balance test" (or "randomisation check")? What is its purpose and its limitations?

#### A balance test (or randomization check) is a statistical test used to assess whether pre-treatment characteristics are similar across treatment and control groups. Its purpose is to verify that randomization was successful in creating comparable groups, ensuring that any observed differences in outcomes are due to the treatment rather than pre-existing differences. 

#### A limitation of balance tests is that they can only check observable characteristics; unmeasured confounders may still be imbalanced, leading to biased causal estimates. Additionally, failing a balance test does not necessarily imply that causal inference is invalid, but it suggests that adjustments (e.g., regression controls, matching) may be needed to account for imbalances.

6. What is post-treatment bias (or overcontrol), and why is it a problem?

#### Post-treatment bias (or overcontrol) occurs when a variable that is affected by the treatment is included as a control in a regression or causal analysis. This is a problem because controlling for such a variable can block part of the causal pathway, leading to biased or misleading estimates of the treatment effect. By adjusting for a post-treatment variable, researchers may remove some of the actual effect of the treatment or introduce spurious relationships that distort causal inference. 

#### For example, if studying the effect of education on income and controlling for job position (which is influenced by education), the estimated effect of education on income may be underestimated.

7. A study examines the impact of a new teaching method on student test scores. Students are randomly assigned to either the new method or a traditional one. However, students in the new method group are more likely to withdraw from the study. What type of selection bias is this, and how might it affect the results?

#### This is an example of attrition bias (a form of selection bias) because students in the new teaching method group are more likely to withdraw from the study, leading to systematic differences in who remains in the sample. If the students who drop out are those who struggle the most with the new method, the remaining students may appear to perform better than they actually would in a full sample, overestimating the effectiveness of the new method. 

#### Conversely, if weaker students remain in the study while stronger ones leave, the treatment effect may be underestimated. Attrition bias threatens the validity of causal estimates because the final comparison between groups no longer reflects the original random assignment.

8. Consider a new educational program and its effect on student test scores. Students are randomly assigned to either the program or a control group. However, the program's effect varies depending on students' prior knowledge. What is this phenomenon called, and how might it affect the interpretation of the results?

#### This phenomenon is called heterogeneous treatment effects, which means that the impact of the educational program differs across students based on their prior knowledge. This variation can affect the interpretation of results because a single average treatment effect may mask important differences in how different subgroups respond to the program. 

#### For example, if students with high prior knowledge benefit more than those with low prior knowledge, the program may appear effective overall but may not help the students who need it most. To fully understand the impact, researchers should analyze treatment effects by subgroups or use interaction terms in regression models to capture these differences.

9. Suppose a new drug's effect on blood pressure is being investigated. Participants are randomly assigned to either the drug or a placebo. However, some participants in the drug group do not adhere to the prescribed dosage. What is this issue called, and how might it affect the results?

#### This issue is called noncompliance, which occurs when participants do not follow the assigned treatment protocol. Noncompliance can affect the results by diluting the estimated treatment effect because the treatment group includes individuals who did not fully receive the intended intervention. 

#### As a result, the comparison between the treatment and control groups may underestimate the true effect of the drug. To address this issue, researchers can use methods such as intention-to-treat analysis, which preserves randomization, or instrumental variable analysis to estimate the effect among those who fully comply with the treatment.


10.  Explain how to estimate the average treatment effect using regression analysis, and why it is a good idea to include covariates in the analysis.

#### To estimate the average treatment effect using regression analysis, researchers typically use an ordinary least squares (OLS) regression model of the form:

#### Yi = B0+B1Ti+ei

#### where Yi is the outcome variable, Ti is a binary indicator for treatment (1 if treated, 0 if control), and B1 represents the estimated average treatment effect. This approach provides an unbiased estimate of the treatment effect if treatment assignment is random.Including covariates in the regression, such as pre-treatment characteristics (Xi), improves precision and controls for any residual differences between groups:

#### Yi = B0+B1T1+B2Xi+ei

#### Adding covariates reduces variance in the outcome, increasing statistical power, and can account for small imbalances that may remain after randomization. However, covariates should be pre-treatment characteristics to avoid post-treatment bias.
